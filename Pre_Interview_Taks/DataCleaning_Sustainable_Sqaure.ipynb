```
#!ls /content/drive/Colab_Notebooks
data = pd.read_csv('/content/drive/Colab_Notebooks/Raw_data.csv',header=None)
```

```
data.head(10)
```

```


new_header = data.iloc[0] #grab the first row for the header
data = data[1:] #take the data less the header row
data.columns = new_header #set the header row as the df header

```

```
data.head(2)

#data.shape
```

```
data.drop_duplicates()
data.shape
```

```
list(data.columns) 
```

```
data = data.loc[:,~data.columns.duplicated()]
```

```
data.shape #Removed 14 duplicate columns
```

```
list(data.columns) 
```

```
data['Response ID'].is_unique
```

```
#Need to find and drop more duplicate columns after analyzing 
#Drop 'Submitted Date/Time', 'Synced Date/Time'(Already captured by 'Submitted Time (Asia/Calcutta)','Synced Time (Asia/Calcutta)'),'Total'(already captured by How many members are there in your family?)

data["Unique"] = data["Total"][~data["Total"].isin(data["How many members are there in your family?"])].drop_duplicates()
data["Unique"] = data["Submitted Date/Time"][~data["Submitted Date/Time"].isin(data["Submitted Time (Asia/Calcutta)"])].drop_duplicates()
data["Unique"] = data["Synced Date/Time"][~data["Synced Date/Time"].isin(data["Synced Time (Asia/Calcutta)"])].drop_duplicates()
data.head(2)
#After analyzing results we can drop 'Submitted Date/Time', 'Synced Date/Time', 'Total'
```

```
data['Gender'].replace('F', 'Female',inplace=True)
data['Gender'].replace('M', 'Male',inplace=True)
data['Gender'].replace('f', 'Female',inplace=True)
data['Gender'].replace('m', 'Male',inplace=True)
data['Gender'].replace('ma', 'Male',inplace=True)
```

```

p_by_state = data.groupby("Gender")["State Wise"].count()
p_by_state.head(10)

#Need to convert all male,female indicators to uniform values
```
